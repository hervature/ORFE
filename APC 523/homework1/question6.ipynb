```python
#  Question 6

import numpy as np
import matplotlib.pyplot as plt

for i in range(45,55):
    iterations = i
    x = np.linspace(1,10,10001)
    y = np.linspace(1,10,10001)

    for i in range(iterations):
        y = np.sqrt(y)

    for i in range(iterations):
        y = np.square(y)

    plt.plot(x, y)
    plt.plot(x,x)
    plt.title('Number of Iterations=' + str(iterations))
    plt.show()
```

```python
# No rounding, just truncating

x = 5
val = [5]

for i in range(54):
    x = np.sqrt(x)
    val.append(x)
    
print(val)
plt.plot(list(range(55)), val)
plt.show()
```

```python
# Close values

iterations = 52
x = np.linspace(1,10,10001)
y = np.linspace(1,10,10001)

for i in range(iterations):
    y = np.sqrt(y)

for i in range(iterations):
    y = np.square(y)
    
print('Values of x that are close: ')
print(x[x - y < 0.00085])

plt.plot(x, y)
plt.plot(x,x)
plt.title('Number of Iterations=' + str(iterations))
plt.show()

# These values are significant because they are the 2nd smallest and 3rd smallest numbers
print(2.719**(1/2**52))
print(7.3891**(1/2**52))
```
